```python
sns.countplot(x=df["pizza_category"], palette='coolwarm')
plt.xlabel("pizza_category")
plt.ylabel("Values")
plt.show()
```

```python
sns.countplot(x=df["pizza_size"])
plt.title("Countplot_of_pizzasize")
plt.show()
```

```python
grp=df.groupby("pizza_category")["quantity"].sum()
```

```python
grp
```

```python
plt.plot(grp,c="red",marker="o")
plt.xlabel("pizza_category")
plt.ylabel("quantity")
plt.show()
```

```python
pvt=pd.pivot_table(data=df, index=["pizza_name","pizza_category"],aggfunc={"quantity":np.sum})
pvt
```

```python
pvt=pd.pivot_table(data=df, index=["pizza_category"],aggfunc={"quantity":np.sum})
pvt
```

```python
plt.plot(pvt,c="red",marker="o")
plt.xlabel("pizza_category")
plt.ylabel("quantity")
```

```python
plt.hist(df["pizza_category"],bins=4, color='g')
plt.title("Histogram_plot")
plt.show()
```

```python
sns.countplot(x=df["pizza_name"]).set_xticklabels(sns.countplot(df['pizza_name']).get_xticklabels(), rotation=90, ha="right")
plt.show()
```

```python
df.head(10)
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x="pizza_category", y="total_price", data=df, width=0.66, saturation=0.65 )
plt.show()
```

```python
pivot=df.pivot_table(index="pizza_size",columns="pizza_category", aggfunc={"total_price":np.mean}, fill_value=0)
pivot
sns.heatmap(pivot, cmap="crest", annot=True,  fmt="2f")
plt.title("Heat map")
plt.show()
```

```python
pvt1=pd.pivot_table(data=df, index="pizza_size", columns="pizza_category",aggfunc={"quantity":np.mean}, fill_value=0)
sns.heatmap(pvt1, annot=True, fmt="2f" )
plt.title("Heat map")
plt.show()
```
